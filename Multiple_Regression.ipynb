```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:\\Users\\pk221\\Downloads\\study.csv")

# Prepare the data
X = data[['Age', 'Experience', 'Hours_Studied']]
y = data['Test_Score']

# Add a column of ones to X to account for the intercept term (b0)
X = np.c_[np.ones(X.shape[0]), X]  # Adds a column of 1s to X

# Compute the coefficients using the normal equation
X_T = X.T  # Transpose of X
X_T_X = X_T.dot(X)  # X^T * X
X_T_X_inv = np.linalg.inv(X_T_X)  # Inverse of (X^T * X)
X_T_y = X_T.dot(y)  # X^T * y
B = X_T_X_inv.dot(X_T_y)  # B = (X^T * X)^-1 * X^T * y

# Predict the values using the computed coefficients
y_pred = X.dot(B)

# Plotting the actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred, color='blue', alpha=0.7, label='Predicted vs Actual')

# Plot the perfect fit line (y = x)
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', lw=2, label='Perfect Fit Line')

# Labels and title
plt.title('Actual vs Predicted Test Scores')
plt.xlabel('Actual Test Scores')
plt.ylabel('Predicted Test Scores')
plt.legend()

# Show the plot
plt.show()

# Plot the regression line for each predictor
# Age vs Test Score
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], y, color='blue', label='Actual Test Score')
plt.plot(data['Age'], B[0] + B[1]*data['Age'], color='red', label='Regression Line')
plt.title('Age vs Test Score')
plt.xlabel('Age')
plt.ylabel('Test Score')
plt.legend()
plt.show()

# Experience vs Test Score
plt.figure(figsize=(10, 6))
plt.scatter(data['Experience'], y, color='green', label='Actual Test Score')
plt.plot(data['Experience'], B[0] + B[2]*data['Experience'], color='red', label='Regression Line')
plt.title('Experience vs Test Score')
plt.xlabel('Experience (Years)')
plt.ylabel('Test Score')
plt.legend()
plt.show()

# Hours Studied vs Test Score
plt.figure(figsize=(10, 6))
plt.scatter(data['Hours_Studied'], y, color='purple', label='Actual Test Score')
plt.plot(data['Hours_Studied'], B[0] + B[3]*data['Hours_Studied'], color='red', label='Regression Line')
plt.title('Hours Studied vs Test Score')
plt.xlabel('Hours Studied per Week')
plt.ylabel('Test Score')
plt.legend()
plt.show()
```
